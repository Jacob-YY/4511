```python
# Initialize empty lists
respondents = []
from_columns = []
one_point = []
minus_one_point = []
zero_one =[]


# Iterate over index(actually the respondent's ID) and corresponding row value
for index, row in res.iterrows():
    
    # Get corresponding column names
    for column in res.columns:
        
        # Initialize a variable one_po to record the options that got score of 1
        one_po = 0 
        
        # Initialize a variable minus_one to record the options that got score of -1
        minus_one = 0 
        
        # Initialize a variable
        zero_three_1 = 0

        
        # Search for a row with score of 1
        if row[column] == 1.0: 
            # If so, assign the option number to one_po
            one_po = int(column.split('_')[1]) 
        
        # Search for a row with score of -1
        elif row[column] == -1.0:
            # If so, assign the option number to minus_one
            minus_one = int(column.split('_')[1]) 
            
        # Search for a row with score of 0
        elif row[column] == 0.0:
            if zero_three_1 == 0:
                zero_three_1 = int(column.split('_')[1])
          
        
        # As long as either one_po or minus_one is not 0
        if one_po or minus_one or zero_three_1:
            
            # Transform the correponding splitted number to a string with info of set number
            from_col = f"{column.split('_')[0]}_DO" 
            
            # Add new respondent ID to the list
            respondents.append(index)
            
            # Add new set number to the list
            from_columns.append(from_col)
            
            # Add new question number that received 1 point to the list
            one_point.append(one_po) 
            
            # Add new question number that received -1 point to the list
            minus_one_point.append(minus_one)
            
            # Add new question number that received 0 point to the list
            zero_one.append(zero_three_1)

# Create a new dataframe by the four lists appended before with processed values
new_df = pd.DataFrame({
    'Respondent': respondents,
    'Question': from_columns,
    'one_point': one_point,
    'minus_one_point': minus_one_point,
    'zero_one':zero_one
})
new_df
```

# Yanhua filling the missing values

```python
result_f = pd.read_excel("result.xlsx", index_col=0)
```

```python
result_f.columns = result_f.columns.str.split('.').str[-1]
result_f.rename(columns={'NaN': 'Respondent'},inplace = True)
res2=result_f.drop(columns=[ '1_DO','2_DO','3_DO','4_DO','5_DO','6_DO','7_DO','8_DO','9_DO','10_DO','11_DO','12_DO'])
df= res2.replace(to_replace = np.nan, value ="0")  
```

```python
df
```

```python
respondents = []
from_columns = []
one_points = []
neg_one_points = []

for index, row in df.iterrows():
    for col in df.columns:
        one_point = 0
        neg_one_point = 0
        
        if row[col] == 1.0:
            one_point = int(col.split('_')[1])
        elif row[col] == -1.0:
            neg_one_point = int(col.split('_')[1])
        
        if one_point or neg_one_point:
            from_col = f"{col.split('_')[0]}"
            respondents.append(index)
            from_columns.append(from_col)
            one_points.append(one_point)
            neg_one_points.append(neg_one_point)
new_df = pd.DataFrame({
    'Respondent': respondents,
    'Question': from_columns,
    'one_point': one_points,
    'ne_one_point': neg_one_points
})
df_combined = new_df.groupby(['Respondent', 'Question']).sum().reset_index()

df_combined = df_combined.set_index('Respondent')
df_combined = df_combined.sort_index()
df_combined.reset_index(inplace=True)
```

```python
# Group the DataFrame by the -1 point column
groupb = df_combined.groupby(['ne_one_point']).size().reset_index(name='count')

sorted_groupb = groupb[groupb['ne_one_point'] != 0]

# Sort the DataFrame by the count -1 point in descending order
sorted_groupb = sorted_groupb.sort_values(by='count', ascending=False)

# Select the top 5 values in the 'ne_one_point' and convert them to a list
ne_one_points_5 = sorted_groupb['ne_one_point'].head(5).tolist()
```

```python
df =df_combined[(df_combined['ne_one_point'] != 0)]
df_min_val = pd.DataFrame(0, index=df['Respondent'].unique(), columns=range(1, 31))

#Group by Respondent and create a DataFram that contains lists of one_point values for each Respondent
list_one_point = df_combined.groupby('Respondent')['one_point'].apply(list).reset_index(name='list_one')

def find_min(row):
    if row['one_point'] in df_min_val.columns:
        df_min_val.at[row['Respondent'], row['one_point']] += 1
    if row['ne_one_point'] in df_min_val.columns:
        df_min_val.at[row['Respondent'], row['ne_one_point']] -= 1

df.apply(find_min, axis=1)

# finds the columns with the minimum value in each row of df_min_val
def find_min_list(row):
    min_value = row.min()
    min_list = []
    for index, value in row.items():
        if value == min_value:
            min_list.append(index)
    return min_list

df_min_val['min_list'] = df_min_val.apply(find_min_list, axis=1)
df_min_val.reset_index(inplace=True)
df_min_val.rename(columns={'index': 'Respondent'},inplace = True)
df_min_val = df_min_val.merge(list_one_point, on='Respondent', how='left')

#removes values from 'min_list' that are not present in 'list_one'
def remove_one_point(row):
    min_list = []
    for min_val in row['min_list']:
        if min_val not in row['list_one']:
            min_list.append(min_val)
    return min_list

df_min_val['min_list'] = df_min_val.apply(remove_one_point, axis=1)

#first attempts to find a minimum value in min_list that matches in the top 5 values in the ne_one_point.
#If not,it selects a random value from 'min_list' 

def get_min_val(row):
    find_min = []
    for val in row['min_list']:
        if val in ne_one_points_5:
            find_min.append(val)
    
    if find_min:
        return find_min[0]
    else:
        return np.random.choice(row['min_list'])
    
df_min_val['min'] = df_min_val.apply(get_min_val, axis=1)

df_min_val = df_min_val[['Respondent', 'min']]
```

```python
tem = pd.merge(df_combined[['Respondent', 'Question', 'one_point', 'ne_one_point']],
               df_min_val[['Respondent', 'min']].rename(columns={'min': 'min_val'}),
               on='Respondent', how='left')

tem['negative_one'] = tem.apply(lambda x: x['min_val'] if x['ne_one_point'] == 0 else x['ne_one_point'], axis=1)

table = tem[['Respondent', 'Question', 'one_point', 'negative_one']]
```

```python
result = []

for (respondent, question), group in table.groupby(['Respondent', 'Question']):
    data = {'Respondent': respondent, 'Question': question}
    
    points = [0] * 30
    
    for idx, row in group.iterrows():
        if row['one_point'] <= 30:
            points[int(row['one_point']) - 1] = 1
        if row['negative_one'] <= 30:
            points[int(row['negative_one']) - 1] = -1
    
    data.update({str(i+1): points[i] for i in range(30)})
    result.append(data)

df_stacked_layout = pd.DataFrame(result)

cols = ['Respondent', 'Question'] + [str(i) for i in range(1, 31)]
df_stacked_layout = df_stacked_layout[cols]
df_stacked_layout=df_stacked_layout.set_index('Respondent')
df_stacked_layout=df_stacked_layout.sort_index()
```
