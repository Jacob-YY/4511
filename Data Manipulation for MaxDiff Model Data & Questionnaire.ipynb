```python
import pandas as pd
import numpy as np
```

#### Data Manipulation for MaxDiff Model Data.csv File

```python
df = pd.read_csv("MaxDiff Model Data3.csv", delimiter=';')
```

```python
# Fill all none values with zero
df.fillna(0,inplace = True) 
df.to_csv('MaxDiff Model Data.csv', index=False)
```

```python
# Show the result
df 
```

```python
# Know specific column names
column_names = df.columns
print(column_names)
```

#### Get rid of unnecessary information and combine the useful groups for different variables

```python
df['CommercialProductName'] = df['CommercialProductName'].str.split('(').str[0]
df['CommercialProductName'] = df['CommercialProductName'].str.split('_').str[0]
df.loc[df['CommercialProductName'] == '25Mbp/s, Kabel TV & bellen ','CommercialProductName'] = '25Mbp/s & Kabel TV & bellen '
grouped = df.groupby(['CommercialProductName'])
count_per_name = grouped.size()
print(count_per_name)
```

```python
df.loc[df['DTV_TIER_OPTION_NAME'] == 'Cable TV (Non Harmo)','DTV_TIER_OPTION_NAME'] = 'Cable TV'
grouped = df.groupby(['DTV_TIER_OPTION_NAME'])
count_per_name = grouped.size()
print(count_per_name)
```

```python
df.loc[df['MediaBoxType'] == 'MEDIABOX XL (SELENE)','MediaBoxType'] = 'MEDIABOX XL'
grouped = df.groupby(['MediaBoxType'])
count_per_name = grouped.size()
print(count_per_name)
```

```python
df.loc[df['Tenure'] == 'Ca.ble. TV','Tenure'] = 'Cable TV'
df.loc[df['Tenure'] == 'ME.DIA.BOX. XL','Tenure'] = 'MEDIABOX XL'
grouped = df.groupby(['Tenure'])
count_per_name = grouped.size()
print(count_per_name)
```

```python
df.loc[df['TVContentPack'] == 'Cable TV (Non Harmo)','TVContentPack'] = 'Cable TV'
grouped = df.groupby(['TVContentPack'])
count_per_name = grouped.size()
print(count_per_name)
```

```python
df.rename(columns={'Sport,': 'Sport'},inplace = True)
df.loc[df['Sport'] == '0, ', 'Sport'] = '0'
df.loc[df['Sport'] == '0,', 'Sport'] = '0'
df.loc[df['Sport'] == '1,', 'Sport'] = '1'
grouped = df.groupby(['Sport'])
count_per_name = grouped.size()
print(count_per_name)
```
